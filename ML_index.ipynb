## Overview of Machine Learning Tutorial

Through machine learning (ML) tutorials, you will understand the concept and purpose of each techique of data analysis and ML modeling and practice them using colab. The tutorials will be broken down into three main 
parts as followed.



1. Data analysis

  a. Data handling
  
  b. Data indexing (numpy, pandas)
  
  c. 2D graph plotting (matplotlib)

  b. Signal processing (FFT, STFT, WT)
  
  e. Feature extraction & selection (t-Test)

  f. Data visualization (PCA, t-SNE)

2. Machine learning

  a. Support vector machine (SVM) 
  
  b. K-nearest neighbor (KNN)

  c. Artificial neural network (ANN)

  d. Hyperparameter optimization (grid search)

  e. Multi-classification (3 class)

3. Deep learning

  a. Convolutional neural network (CNN)

  b. Recurrent neural network (RNN)
