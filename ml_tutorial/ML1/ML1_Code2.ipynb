#### 1-1-3. Generate Array Data by Designating a Range

np.arange(a) : Generate an array greater than or equal to 0 and less than 'a'

```python
ary_range_1 = np.arange(5)
ary_range_1
```

np.arange(a,b) : Generate an array greater than or equal to 'a' and less than 'b'

```python
ary_range_2 = np.arange(1,10)
ary_range_2
```

np.arange(a,b,c) : Generate an array greater than or equal to 'a' and less than 'b' with an interval 'c'

```python
ary_range_3 = np.arange(1,10,0.5)
ary_range_3
```

#### Range Definition of Basic Python

range(a) : Generate a range greater than or equal to 0 and less than 'a'
* It is not an array data. (mainly applying it for loops)

```python
range_1 = range(5)
range_1
```

range(a,b) : Generate a range greater than or equal to 'a' and less than 'b'

```python
range_2 = range(1,10)
range_2
```
