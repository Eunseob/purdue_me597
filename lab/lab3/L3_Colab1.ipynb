## 1.3 Power Meter Data Collection Using the Modbus Protocol

The details of the power meter system and hardware configuration are shown in Figure 6. The power meter is equipped with Modbus RTU communication option. Because Modbus RTU which uses serial communications cannot have multiple masters, we will use Modbus RTU (slave side) to Modbus TCP (master side) converter (Moxa, [MGate MB3180](https://www.moxa.com/en/products/industrial-edge-connectivity/protocol-gateways/modbus-tcp-gateways/mgate-mb3180-mb3280-mb3480-series)). The converter allows us to communicate the Modbus RTU device via Modbus TCP. Modbus RTU is limited to a single master with multiple slaves whereas multiple masters and multiple slaves’ configurations are possible if using Modbus TCP. Schematics of Modbus RTU and Modbus TCP are shown in Figure 7. Since Modbus RTU and Modbus TCP share and follow the standard Modbus protocol, it is possible to use converting data each other. In the most industry, Modbus RTU is commonly used due to cost and the reliability of data transmission. The main difference between Modbus RTU and Modbus TCP is mostly in the transport frames or the wrapper around the PDU (Protocol Data Unit). In both forms of Modbus, application specific addressing and error checking are attached to the PDU to make the ADU (Application Data Unit). In Modbus TCP is encapsulated in a TCP packet. The TCP protocol handless the error checking, which is why it is omitted from the Modbus TCP ADU. The comparisons of Modbus data frames are shown in Figure 8.

## 1.4 IO-Link Data Collection Using JSON interface

Unlike traditional control signals, IO-Link offers increased bandwidth and full access to the sensor’s intelligence without changing sensors and wiring. Using your existing architecture, IO-Link opens new possibilities to maximize the overall efficiency of your machines and processes. From easy and fast setup to additional machine monitoring data to process transparency from the machine level to the ERP software, IO-Link offers many benefits. See below for just some othe ways you can use IO-Link.

The IO-Link master provides useful features. It separates between automation network and IT network so that we use the IT network for monitoring purpose. It has 8 ports to be connected to the IO-Link devices such as the vibration sensor. Moreover, it transmits machine data, process parameters and diagnostic data to the controller with ProfiNet (PLC/controller side) interface and separate MQTT JSON interface (IoT/ERP side) as Figure 10. In this lab, we will use JSON interface to extract vibration data from the IO-Link master.

The sensor placement and hardware configuration are shown in Figure 11. The vibration sensor has a single axis. Note that when you deploy vibration/acceleration sensor to a rotary machine, the rule of thumb for placement is to make the measurement axis be a radial direction of the rotating motion.
