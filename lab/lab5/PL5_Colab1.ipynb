When the MTConnect adapter is made a connection to the agent, the EVENT item id ‘avail’ is changed to ‘AVAILABLE’. In reverse, the MTConnect adapter isn’t made a connection to the agent, it is changed to ‘UNAVAILABLE’. Therefore, by checking the ‘Availability’ (id = ‘avail’) we can see whether the adapter is connected to the agent well or not. If the adapter is connected but the SAMPLE values (sensor data) are not changing, we may consider the issues on the sensor data access.
