Because only X-axis acceleration (id = *‘a1’*) and temperature (id = *‘t1’*) values are added as the data items of MTConnect adapter and having the random values, other data items are still showing ‘UNAVAILABLE’. When the MTConnect adapter is made a connection to the agent, the EVENT item id ‘avail’ is changed to ‘AVAILABLE’. In reverse, the MTConnect adapter isn’t made a connection to the agent, it is changed to ‘UNAVAILABLE’. Therefore, by checking the ‘Availability’ (id = ‘avail’) we can see whether the adapter is connected to the agent well or not. If the adapter is connected but the SAMPLE values (sensor data) are not changing, we may consider the issues on the sensor data access. Please note that to halt (forcedly stop) the MTConnect adapter, the shortcut on the terminal is Ctrl + c. And the printing out values is to see what values are transmitting to the agent.
